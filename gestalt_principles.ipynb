```python
data_2015 = pd.read_csv('world-happiness/2015.csv')
data_2015.head()
```

```python
data_2016 = pd.read_csv('world-happiness/2016.csv')
data_2016.head()
```

```python
data_2017 = pd.read_csv('world-happiness/2017.csv')
data_2017.head()
```

```python
data_2018 = pd.read_csv('world-happiness/2018.csv')
data_2018.head()
```

```python
data_2019 = pd.read_csv('world-happiness/2019.csv')
data_2019.head()
```

```python
data_comb = pd.DataFrame()
data_comb['Country'] = data_2015['Country']
data_comb['2015'] = data_2015['Happiness Score']
data_comb = pd.merge(
    data_comb, data_2016.loc[:,['Country', 'Happiness Score']], left_on='Country', right_on='Country'
)
data_comb.rename(columns={'Happiness Score':'2016'}, inplace=True)
data_comb = pd.merge(
    data_comb, data_2017.loc[:,['Country', 'Happiness.Score']], left_on='Country', right_on='Country'
)
data_comb.rename(columns={'Happiness.Score':'2017'}, inplace=True)
data_comb = pd.merge(
    data_comb, data_2018.loc[:,['Country or region', 'Score']], left_on='Country', right_on='Country or region'
)
data_comb.rename(columns={'Score':'2018'}, inplace=True)
data_comb = pd.merge(
    data_comb, data_2019.loc[:,['Country or region', 'Score']], left_on='Country', right_on='Country or region'
)
data_comb.rename(columns={'Score':'2019'}, inplace=True)
data_comb.drop(['Country or region_x', 'Country or region_y'], axis=1, inplace=True)
data_comb.set_index('Country', inplace=True)
data_comb
```

```python
def plot_ci_manual(t, s_err, n, x, x2, y2, ax=None):
    """Return an axes of confidence bands using a simple approach.

    Notes
    -----
    .. math:: \left| \: \hat{\mu}_{y|x0} - \mu_{y|x0} \: \right| \; \leq \; T_{n-2}^{.975} \; \hat{\sigma} \; \sqrt{\frac{1}{n}+\frac{(x_0-\bar{x})^2}{\sum_{i=1}^n{(x_i-\bar{x})^2}}}
    .. math:: \hat{\sigma} = \sqrt{\sum_{i=1}^n{\frac{(y_i-\hat{y})^2}{n-2}}}

    References
    ----------
    .. [1] M. Duarte.  "Curve fitting," Jupyter Notebook.
       http://nbviewer.ipython.org/github/demotu/BMC/blob/master/notebooks/CurveFitting.ipynb

    """
    if ax is None:
        ax = plt.gca()

    ci = t * s_err * np.sqrt(1/n + (x2 - np.mean(x))**2 / np.sum((x - np.mean(x))**2))
    plt.fill_between(x2, y2 + ci, y2 - ci, color="#b9cfe7", edgecolor="")

    return ax


def equation(a, b):
    """Return a 1D polynomial."""
    return np.polyval(a, b)
```

```python
import scipy.stats as stats
n = data_2016['Economy (GDP per Capita)'].size 
                              # used for CI and PI bands
p, cov = np.polyfit(data_2016['Happiness Score'], data_2016['Economy (GDP per Capita)'], 1, cov=True) 
m = p.size 
t = stats.t.ppf(0.975, n - m)
y_model = equation(p, data_2016['Happiness Score'])  
resid = data_2016['Economy (GDP per Capita)'] - y_model
dof = n - m
s_err = np.sqrt(np.sum(resid**2) / dof) 
   
x2 = np.linspace(np.min(data_2016['Happiness Score']), np.max(data_2016['Happiness Score']), 100)
y2 = equation(p, x2)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(x=data_2016['Happiness Score'], y=data_2016['Economy (GDP per Capita)'], c='#CCCCCC')
b, a = np.polyfit(data_2016['Happiness Score'], data_2016['Economy (GDP per Capita)'], 1)
yfit = [a + b * xi for xi in data_2016['Happiness Score']]
plot_ci_manual(t, s_err, n, data_2016['Happiness Score'], x2, y2)
plt.plot(data_2016['Happiness Score'], yfit, c='r')
plt.title("More money = more happy?", fontsize="18", fontweight="bold")
plt.xlabel("Happiness Score", fontsize="18")
plt.ylabel("Economy (GDP per Capita)", fontsize="18")
plt.show()
```

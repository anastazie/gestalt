```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn import datasets
```

```python
# Import example data
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris['target']
df.head()
```

# Simplicity

```python
data1 = [(12, 'A'), (30, 'B'), (1, 'C'), (8, 'D'), (22, 'E')]
```

```python
data1_sorted = sorted(data1, key=lambda tup: tup[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.4

# set height of bar
bars1 = [s[0] for s in data1]


# Set position of bar on X axis
r1 = np.arange(len(bars1))
 
# Make the plot
plt.bar(r1, bars1, color='#0072b2', width=barWidth, edgecolor='white', label='var1')

# Add xticks on the middle of the group bars
plt.xlabel('group', fontweight='bold')
plt.ylabel('Variable', fontweight='bold')
plt.xticks(np.arange(len(bars1)), [s[1] for s in data1])

plt.show()
```

# Proximity

```python
# Dataset:
a = pd.DataFrame({'group' : np.repeat('A',500), 'value': np.random.normal(10, 3, 500)})
b = pd.DataFrame({'group' : np.repeat('B',500), 'value': np.random.normal(18, 1.2, 500)})
c = pd.DataFrame({'group' : np.repeat('C',20), 'value': np.random.normal(25, 4, 20)})
d = pd.DataFrame({'group' : np.repeat('D',100), 'value': np.random.uniform(12, size=100)})
box_data1 = a.append(b).append(c).append(d)

# Usual boxplot
sns.boxplot(x='group', y='value', data=box_data1)
plt.xlabel('Group', fontweight='bold')
plt.ylabel('Value', fontweight='bold')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for i, t in enumerate(df['target'].unique()):
    ax.scatter(
        df[df['target'] == t]['sepal length (cm)'], df[df['target'] == t]['sepal width (cm)'], 
        alpha=0.8, c='#808080', edgecolors='none', s=30
    )
plt.xlabel('Sepal length (cm)', fontweight='bold')
plt.ylabel('Sepal width (cm)', fontweight='bold')
plt.show()
```

```python
markers = [ "," , "o" , "v" , "^" , "<", ">"]
colors = ('#cc79a7', '#0072b2', '#009e73')
groups = iris.target_names

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for i, t in enumerate(df['target'].unique()):
    ax.scatter(
        df[df['target'] == t]['sepal length (cm)'], df[df['target'] == t]['sepal width (cm)'], 
        alpha=0.8, c=colors[i], edgecolors='none', s=30, label=groups[i], marker=markers[i]
    )
plt.xlabel('Sepal length (cm)', fontweight='bold')
plt.ylabel('Sepal width (cm)', fontweight='bold')
plt.legend(loc=1)
plt.show()
```

# Similarity

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(1, 1, 1)
for i, t in enumerate(df['target'].unique()):
    ax.scatter(
        df[df['target'] == t]['sepal length (cm)'], df[df['target'] == t]['sepal width (cm)'], 
        alpha=0.8, c='#808080', edgecolors='none', s=50
    )
plt.plot(df.iloc[df['sepal width (cm)'].idxmax()]['sepal length (cm)'],
         df.iloc[df['sepal width (cm)'].idxmax()]['sepal width (cm)'], marker='o', markersize=15, color='#d55e00') 
plt.xlabel('Sepal length (cm)', fontweight='bold')
plt.ylabel('Sepal width (cm)', fontweight='bold')
plt.show()
```

# Figure and ground

```python
markers = [ "," , "o" , "v" , "^" , "<", ">"]
colors = ('#cc79a7', '#0072b2', '#009e73')
groups = iris.target_names

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for i, t in enumerate(df['target'].unique()):
    ax.scatter(
        df[df['target'] == t]['sepal length (cm)'], df[df['target'] == t]['sepal width (cm)'], 
        alpha=0.8, c=colors[i], edgecolors='none', s=30, label=groups[i], marker=markers[i]
    )
plt.xlabel('Sepal length (cm)', fontweight='bold')
plt.ylabel('Sepal width (cm)', fontweight='bold')
ax.patch.set_facecolor('black')
plt.legend(loc=1)
plt.show()
```
